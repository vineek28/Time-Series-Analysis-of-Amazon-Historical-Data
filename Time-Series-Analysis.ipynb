The Pattern: If the price is stationary, then when the price goes "too high," the next move is likely "down" toward the mean. This negative feedback loop is a repeatable pattern.
The "Identification" Moment: When the p-value drops below 0.05, the math is essentially saying: "The probability that this data is just wandering randomly is less than 5%. We have identified a consistent, repeating structure."

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['Returns'])
print(result[0])
print(result[1])
```

This is your ADF Statistic. Think of this as a measure of "intensity."
The ADF test searches for a "unit root" (the mathematical signature of a random walk/non-stationary data).
The more negative this number is, the stronger the evidence against non-stationarity.
A value of -14.4 is extremely large (in absolute terms). For context, usually, any value more negative than -3.5 is considered a very strong "hit."
2. The P-Value: 8.39e-27
This is the most important number for identification. Written out, it looks like this:
0.00000000000000000000000000839
This represents the probability that your data is actually just a random walk and you just got "lucky" seeing a pattern.
In science and finance, we usually look for a p-value less than 0.05 (5% chance of being wrong).

```python
import numpy as np 
split_index = int(len(df)*0.8)
train = df.iloc[:split_index]
test = df.iloc[split_index:]
```

```python
#forecasting of naive mean

import numpy as np 
mean_returns = train["Returns"].mean()
naive_forecast_model = np.full(len(test), mean_returns)
```

```python
print(mean_returns)
print(naive_forecast_model)
```

```python
from sklearn.metrics import mean_squared_error

naive_rmse = np.sqrt(mean_squared_error (test['Returns'], naive_forecast_model))
```

```python
print(naive_rmse)
```

thenically the naive rmse represents:
The average prediction error if we simply predict tomorrow‚Äôs return as the historical mean.

this is our benchmark error.

IN THE END WHAT DOES THIS MEAN? 

The Logical Conflict: "Stationary but Unpredictable"
This is why stationarity alone isn't a "money-making" signal.
Stationarity means the data stays within a boundary (it won't fly off to infinity).
Autocorrelation means there is a "link" between yesterday and today.
If your PACF is flat, it means you have successfully removed the trend, but what is left behind is "Pure Randomness." In finance, this is known as the Efficient Market Hypothesis. It suggests that the "Signal" you were looking for is already fully "priced in," leaving only random, unpredictable noise behind.


So, it means we cannot predict this using only returns. so we need to opt for non linear models.
